## Step 1: Understanding the Lotka-Volterra Model
The Lotka-Volterra equations model the interaction between predator and prey populations:
- dx/dt = αx - βxy (prey growth and predation)
- dy/dt = δxy - γy (predator growth and death)

Where:
- x is prey population (e.g., rabbits)
- y is predator population (e.g., foxes)
- α: prey growth rate
- β: predation rate
- δ: predator growth rate from predation
- γ: predator death rate

## Step 2: Setting up the environment
Let me import the necessary packages for solving and plotting the differential equations:

</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

<div style="display: flex; align-items: center; margin-bottom: 12px;">
    <h3 style="color: #1d5b8e; margin: 0; margin-right: 12px; font-size: 14px; font-weight: 600;">Assistant</h3>
    <div style="flex: 1; height: 2px; background-color: #1d5b8e;"></div>
</div>
<div style="margin-bottom: 16px;">## Step 3: Defining the Lotka-Volterra equations
Now I'll define the system of differential equations as a function that can be used by the ODE solver:

</div>

```python
def lotka_volterra(z, t, alpha, beta, delta, gamma):
    """
    Lotka-Volterra equations for predator-prey dynamics
    
    Parameters:
    z: array [x, y] where x is prey, y is predator
    t: time
    alpha, beta, delta, gamma: model parameters
    
    Returns:
    dzdt: derivatives [dx/dt, dy/dt]
    """
    x, y = z
    dxdt = alpha * x - beta * x * y
    dydt = delta * x * y - gamma * y
    return [dxdt, dydt]
```

<div style="display: flex; align-items: center; margin-bottom: 12px;">
    <h3 style="color: #1d5b8e; margin: 0; margin-right: 12px; font-size: 14px; font-weight: 600;">Assistant</h3>
    <div style="flex: 1; height: 2px; background-color: #1d5b8e;"></div>
</div>
<div style="margin-bottom: 16px;">## Step 4: Setting parameters and initial conditions
Let me define realistic parameters and initial populations for the model:

</div>

```python
# Model parameters
alpha = 1.0   # prey growth rate
beta = 0.1    # predation rate
delta = 0.075 # predator growth rate from predation
gamma = 1.5   # predator death rate

# Initial conditions
x0 = 40  # initial prey population
y0 = 9   # initial predator population
z0 = [x0, y0]

# Time points
t = np.linspace(0, 15, 1000)

print(f"Parameters: α={alpha}, β={beta}, δ={delta}, γ={gamma}")
print(f"Initial conditions: Prey={x0}, Predator={y0}")
```

<div style="display: flex; align-items: center; margin-bottom: 12px;">
    <h3 style="color: #1d5b8e; margin: 0; margin-right: 12px; font-size: 14px; font-weight: 600;">Assistant</h3>
    <div style="flex: 1; height: 2px; background-color: #1d5b8e;"></div>
</div>
<div style="margin-bottom: 16px;">## Step 5: Solving the differential equations
Now I'll solve the system using scipy's odeint solver:

</div>

```python
# Solve the ODEs
solution = odeint(lotka_volterra, z0, t, args=(alpha, beta, delta, gamma))
x, y = solution.T

print(f"Solution shape: {solution.shape}")
print(f"Final prey population: {x[-1]:.2f}")
print(f"Final predator population: {y[-1]:.2f}")
```

<div style="margin-bottom: 16px;">## Step 6: Creating a clean, professional plot
Now I'll create a visually appealing plot showing both the time series and the phase diagram:

</div>

```python
# Set up the figure with subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot time series
ax1.plot(t, x, label='Prey (Rabbits)', color='#2E86AB', linewidth=2.5)
ax1.plot(t, y, label='Predator (Foxes)', color='#A23B72', linewidth=2.5)
ax1.set_xlabel('Time', fontsize=12)
ax1.set_ylabel('Population', fontsize=12)
ax1.set_title('Lotka-Volterra Predator-Prey Dynamics', fontsize=14, fontweight='bold')
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)
ax1.set_xlim(0, 15)

# Plot phase diagram
ax2.plot(x, y, color='#F18F01', linewidth=2.5)
ax2.scatter(x[0], y[0], color='green', s=100, label='Initial state', zorder=5)
ax2.scatter(x[-1], y[-1], color='red', s=100, label='Final state', zorder=5)
ax2.set_xlabel('Prey Population', fontsize=12)
ax2.set_ylabel('Predator Population', fontsize=12)
ax2.set_title('Phase Diagram', fontsize=14, fontweight='bold')
ax2.legend(fontsize=11)
ax2.grid(True, alpha=0.3)

# Improve overall appearance
plt.tight_layout()

# Add a main title
fig.suptitle('Predator-Prey Population Model', fontsize=16, fontweight='bold', y=1.02)

# Show the plot
plt.show()
```
